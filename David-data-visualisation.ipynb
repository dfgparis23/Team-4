### Utility functions

```python
def plotFrequencies(df, figsize=None):
    rows = (len(df.columns) + 1) // 2
    fig, axes = plt.subplots(nrows=rows,ncols=2, figsize=figsize)
    for i, column in enumerate(df.columns):
        df[column].value_counts().plot(ax=axes[i // 2, i % 2], kind='bar', subplots=True)
    plt.show()
```
